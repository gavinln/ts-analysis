## Multiple time series

### Multivariate time series

Two or more variables over time with variables representing different kinds of measurements

* Annotation with multiple variables
* Forecasting with exogenous variables
* Vector forecasting (forecasting multiple series at the same time)

```python
temperature = utils.load_temperature()
fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(16, 8), sharex=True)
pressure.plot(ax=ax0)
temperature.plot(ax=ax1)
ax0.set_ylabel('Pressure')
ax1.set_ylabel('Temperature')
ax1.set_xlabel('Time')
```

### Panel data

Also called longitudinal data, two or more instances of the same kind of measurement 

* time series annotation
* Forecasting
* Series as a feature learning tasks

In Panel data each instance is assumed to be independent

```python
experiments = utils.load_experiments(variables="pressure")
```

```python
fig, ax = plt.subplots(figsize=(16, 4))
experiments.sample(10).T.plot(ax=ax)
ax.set_ylabel("Pressumre")
ax.set_xlabel("Time")
```

Panel data may be multi-variate. Panel data may also have time-constant variables

## Time series data and statistical dependence

* Time series observations are statistically dependent on past observations
* In multivariate data it is implausible to assume different variables are independent
* In panel data it is plausible to assume different variables are independent
